```python
import pandas as pd

test_video_path = 'deepfake-detection-challenge/test_videos/'
train_video_path = 'deepfake-detection-challenge/train_sample_videos/'

df = pd.read_csv('deepfake-detection-challenge/sample_submission.csv')
df
```

```python
# reading a video
import cv2
import numpy as np
 
file_name = df['filename'][0]    
print(file_name)
    
# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture(test_video_path+file_name)
 
# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
 
# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frame
  ret, frame = cap.read()
  if ret == True:
 
    # Display the resulting frame
    cv2.imshow('Frame',frame)
 
    # Press Q on keyboard to  exit
    if cv2.waitKey(25) & 0xFF == ord('q'):
      break
 
  # Break the loop
  else: 
    break
 
# When everything done, release the video capture object
cap.release()
 
# Closes all the frames
cv2.destroyAllWindows()
```
